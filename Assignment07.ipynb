```python
for index, row in survey.iterrows():
    first_element = row[0]
    second_element = row[1]
    third_element = row[2]
    fourth_element = row[3]
    fifth_element = row[4]
    sixth_element = row[5]
    survey.at[index, survey.columns[0]] = float(first_element)

    if second_element == "TRUE":
        survey.at[index, survey.columns[1]] = True
    else:
        survey.at[index, survey.columns[1]] = False

    if third_element == "TRUE":
        survey.at[index, survey.columns[2]] = True
    else:
        survey.at[index, survey.columns[2]] = False

    if fourth_element == "TRUE":
        survey.at[index, survey.columns[3]] = True
    else:
        survey.at[index, survey.columns[3]] = False

    if fifth_element == "None":
        survey.at[index, survey.columns[4]] = None
    else:
        survey.at[index, survey.columns[4]] = fifth_element

    survey.at[index, survey.columns[5]] = float(sixth_element)

print(survey)
```

```python
def categorize_years(years):
    if years < 5:
        return "< 5 Years"
    elif years < 10:
        return "5-10 Years"
    elif years < 15:
        return "10-15 Years"
    elif years < 20:
        return "15-20 Years"
    elif years < 25:
        return "20-25 Years"
    else:
        return "25+ Years"

survey['category'] = survey.iloc[:, 0].apply(categorize_years)

print(survey)
```

```python
lessthan5 = []
fiveto10 = []
tento15 = []
fifteento20 = []
twentyto25 = []
moretha25 = []

lessthan5 = survey[survey['category'] == "< 5 Years"].iloc[:, -2].tolist()
fiveto10 = survey[survey['category'] == "5-10 Years"].iloc[:, -2].tolist()
tento15 = survey[survey['category'] == "10-15 Years"].iloc[:, -2].tolist()
fifteento20 = survey[survey['category'] == "15-20 Years"].iloc[:, -2].tolist()
twentyto25 = survey[survey['category'] == "20-25 Years"].iloc[:, -2].tolist()
moretha25 = survey[survey['category'] == "25+ Years"].iloc[:, -2].tolist()
print("Less than 5 years:", lessthan5)
print("5-10 years:", fiveto10)
print("10-15 years:", tento15)
print("15-20 years:", fifteento20)
print("20-25 years:", twentyto25)
print("More than 25 years:", moretha25)
```

```python
categories_dict = {
    "< 5 Years": lessthan5,
    "5-10 Years": fiveto10,
    "10-15 Years": tento15,
    "15-20 Years": fifteento20,
    "20-25 Years": twentyto25,
    "25+ Years": moretha25
}

for category, values in categories_dict.items():
    if values: 
        print(f"Average for {category}: {sum(values) / len(values)}")
    else:
        print(f"Average for {category}: No data")
```

```python
survey.columns = survey.columns.append(pd.Index(["categories"]))
print(survey.columns)
```
